```python
import pandas as pd
import numpy as np
import datetime
```

```python
df = pd.read_csv('./prj_sns_trend_crawling_instagram_test.csv')
df.drop([df.columns[0]],axis=1, inplace=True)
df
```

```python
#NaN값 제거
df.isnull().sum()
df = df.dropna()
df.reset_index(drop =True, inplace = True)

print(df)
```

```python
for i in range(len(df)):
    df['date'][i] = df['date'][i].replace('년 ', '-').replace('월 ', '-').replace('일', '')
df['date']
```

```python
#datetime으로 변경
df['date'] = pd.to_datetime(df['date'])
```

```python
#날짜 포맷 변경
for i in range(len(df)):
    df['date'][i] = df['date'][i].strftime('%Y-%m-%d')
df['date']
```
