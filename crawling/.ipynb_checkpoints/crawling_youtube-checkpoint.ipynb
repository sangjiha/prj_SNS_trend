```python
#유튜브 크롤링은 특성 상, 코드 2개로 나눠서 진행 -> 1번 크롤링 후 test 파일 저장 후 2번 크롤링 진행

#크롬 드라이버 열기 
import time
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from selenium.common.exceptions import NoSuchElementException, ElementNotInteractableException
from selenium import webdriver
from bs4 import BeautifulSoup
from selenium.webdriver.common.keys import Keys
options = webdriver.ChromeOptions()
#options.add_argument('headless')
options.add_argument('disable-gpu')
options.add_argument('lang=ko_KR')
driver = webdriver.Chrome('./chromedriver', options=options)
body = driver.find_element_by_tag_name('body')

title = [] #크롤링할 컬럼들 
date = []
youtuber_name = []
comments = []
likes = []

url = 'https://www.youtube.com/results?search_query=%EC%B9%98%EC%A6%88%EB%B3%BC+&sp=EgIIBQ%253D%253D' #치즈볼 올해 검색 링크
driver.get(url)
time.sleep(0.4)
```

```python
#크롤링 1번 
for j in range(1,20): #영상 xpath 1~20 변화
        try:
            driver.find_element_by_xpath('/html/body/ytd-app/div/ytd-page-manager/ytd-search/div[1]/ytd-two-column-search-results-renderer/div/ytd-section-list-renderer/div[2]/ytd-item-section-renderer/div[3]/ytd-video-renderer[{0}]/div[1]/div/div[1]/div/h3/a/yt-formatted-string'.format(j)).click()
            body = driver.find_element_by_tag_name('body')
            body.send_keys(Keys.PAGE_DOWN) #스크롤 내리기 
            time.sleep(1.0)
            body.send_keys(Keys.PAGE_DOWN)
            time.sleep(1.0)
            body.send_keys(Keys.PAGE_DOWN)
            time.sleep(1.0)
            body.send_keys(Keys.PAGE_DOWN)
            time.sleep(2.0)
            
            title_name = driver.find_element_by_xpath('//*[@id="container"]/h1/yt-formatted-string').text
            dates = driver.find_element_by_xpath('//*[@id="date"]/yt-formatted-string').text #날짜
            good = driver.find_elements_by_xpath('/html/body/ytd-app/div/ytd-page-manager/ytd-watch-flexy/div[4]/div[1]/div/div[5]/div[2]/ytd-video-primary-info-renderer/div/div/div[3]/div/ytd-menu-renderer/div/ytd-toggle-button-renderer[1]/a/yt-formatted-string') #좋아요 수 
            good = good[0].get_attribute('aria-label') #좋아요 수
            youtuber = driver.find_element_by_xpath('/html/body/ytd-app/div/ytd-page-manager/ytd-watch-flexy/div[4]/div[1]/div/div[6]/div[3]/ytd-video-secondary-info-renderer/div/div[2]/ytd-video-owner-renderer/div[1]/ytd-channel-name/div/div/yt-formatted-string/a').text #유튜버
            body = driver.find_element_by_tag_name('body')
            body.send_keys(Keys.PAGE_DOWN) #스크롤 내리기 
            
            
            comment = ''
            
            for k in range(2,22): #한 영상 당 댓글 20개씩 크롤링
                
                try:
                    temp_comment = driver.find_element_by_xpath('/html/body/ytd-app/div/ytd-page-manager/ytd-watch-flexy/div[4]/div[1]/div/ytd-comments/ytd-item-section-renderer/div[3]/ytd-comment-thread-renderer[{}]/ytd-comment-renderer/div[1]/div[2]/ytd-expander/div/yt-formatted-string[2]'.format(k)).text

                    if k % 3 == 0:
                        body.send_keys(Keys.PAGE_DOWN) #댓글 스크롤 내리기
                        time.sleep(3.0)
                    comment = comment + (' ' + temp_comment) #댓글 

                    
                except:
                    comment += ' ' #댓글 없는 영상일 경우, 공백 추가 
                    continue
            comments.append(comment) #댓글 
            youtuber_name.append(youtuber) #유튜버
            title.append(title_name) #영상 제목
            likes.append(good) #좋아요 수 
            date.append(dates) #날짜

            
            driver.back() #한 영상 댓글 20개 크롤링 후 뒤로가기
            time.sleep(5.0)
            body = driver.find_element_by_tag_name('body')
            body.send_keys(Keys.PAGE_DOWN) #영상들 내리기 위한 스크롤 다운 
        except:
            print(j)
            continue
```

```python
#크롤링 1번 완료 후 test 파일로 저장 
df = pd.DataFrame({'youtuber':youtuber_name,'date':date,'title':title, 'comment':comments, 'likes':likes})  #크롤링 후 데이터프레임 생성
df.to_csv('./crawling_youtube_test.csv') #크롤링 완료 후 데이터프레임 저장 
```

```python
#크롤링 2번 바로 진행  
for i in range(1,10000): #크롤링 범위 최대로 설정
    
    for j in range(1,20): #20까지 설정하면 됨 
        try:
            driver.find_element_by_xpath('/html/body/ytd-app/div/ytd-page-manager/ytd-search/div[1]/ytd-two-column-search-results-renderer/div/ytd-section-list-renderer/div[2]/ytd-item-section-renderer[{0}]/div[3]/ytd-video-renderer[{1}]/div[1]/div/div[1]/div/h3/a/yt-formatted-string'.format(i,j)).click()
            time.sleep(2.0)
            body = driver.find_element_by_tag_name('body')
            body.send_keys(Keys.PAGE_DOWN) #스크롤 다운
            time.sleep(1.0)
            body = driver.find_element_by_tag_name('body')
            body.send_keys(Keys.PAGE_DOWN)
            time.sleep(1.0)
            body = driver.find_element_by_tag_name('body')
            body.send_keys(Keys.PAGE_DOWN)
            time.sleep(1.0)
            body = driver.find_element_by_tag_name('body')
            body.send_keys(Keys.PAGE_DOWN)
           
            
            title_name = driver.find_element_by_xpath('//*[@id="container"]/h1/yt-formatted-string').text #제목
            dates = driver.find_element_by_xpath('//*[@id="date"]/yt-formatted-string').text #날짜
            good = driver.find_elements_by_xpath('/html/body/ytd-app/div/ytd-page-manager/ytd-watch-flexy/div[4]/div[1]/div/div[5]/div[2]/ytd-video-primary-info-renderer/div/div/div[3]/div/ytd-menu-renderer/div/ytd-toggle-button-renderer[1]/a/yt-formatted-string') #좋아요 수 
            good = good[0].get_attribute('aria-label')
            youtuber = driver.find_element_by_xpath('/html/body/ytd-app/div/ytd-page-manager/ytd-watch-flexy/div[4]/div[1]/div/div[6]/div[3]/ytd-video-secondary-info-renderer/div/div[2]/ytd-video-owner-renderer/div[1]/ytd-channel-name/div/div/yt-formatted-string/a').text #유튜버
            body = driver.find_element_by_tag_name('body')
            body.send_keys(Keys.PAGE_DOWN)
            
            
            comment = ''
            
            for k in range(2,22): #댓글 20개씩
                
                try:
                    temp_comment = driver.find_element_by_xpath('/html/body/ytd-app/div/ytd-page-manager/ytd-watch-flexy/div[4]/div[1]/div/ytd-comments/ytd-item-section-renderer/div[3]/ytd-comment-thread-renderer[{}]/ytd-comment-renderer/div[1]/div[2]/ytd-expander/div/yt-formatted-string[2]'.format(k)).text

                    if k % 3 == 0:
                        body.send_keys(Keys.PAGE_DOWN)
                        time.sleep(3.0)
                    comment = comment + (' ' + temp_comment) #댓글 

                    
                except:
                    comment += ' '
                    continue
            comments.append(comment) #댓글
            youtuber_name.append(youtuber) #유튜버
            title.append(title_name) #영상 제목
            likes.append(good) #좋아요 수 
            date.append(dates) #날짜

            driver.back()
            time.sleep(5.0)
            body = driver.find_element_by_tag_name('body')
            body.send_keys(Keys.PAGE_DOWN)
        except:
            print(i,j,'error')
            continue
            

    if i % 100 == 0:
        df = pd.DataFrame({'youtuber':youtuber_name,'date':date,'title':title, 'comment':comments, 'likes':likes}) #크롤링 100개 마다 저장되도록
        df.to_csv('./crawling_youtube_{}.csv'.format(i))

df = pd.DataFrame({'youtuber':youtuber_name,'date':date,'title':title, 'comment':comments, 'likes':likes}) 
df.to_csv('./crawling_youtube_final.csv') #크롤링 완료 후 final로 저장 
print(df.info())
```
