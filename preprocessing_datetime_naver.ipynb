```python
import time
import pandas as pd
import datetime
import numpy as np

```

```python
#맨위의 read_csv와 맨아래의 to_csv 에서 파일명만 주의해주세요.

df = pd.read_csv('../prj_sns_trend_private/crawling_naver_blog.csv')
df.drop([df.columns[0]],axis=1, inplace=True)
df
```

```python


for z in range(len(df['date'])):
    if '전' in df['date'][z]:


        if '일' in df['date'][z]:
            d = int(df['date'][z][0])
          
            new_date = datetime.datetime.now() - datetime.timedelta(days = d)            
            df['date'][z] = new_date


            
        else:
            new_date2 = datetime.datetime.now()
            df['date'][z] = new_date2
            
        
    elif '어제' in df['date'][z]:
        new_date3 = datetime.datetime.now() - datetime.timedelta(days = 1)
        df['date'][z] = new_date3

    
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
for z in range(len(df)):

    df['date'][z] = df['date'][z].strftime('%Y-%m-%d')
```
