```python
import pandas as pd
```

```python
df1 = pd.read_csv('./crawling_data/crawling_instagram_final.csv')
df1.drop(df1.columns[0], axis=1, inplace=True)
df1.drop_duplicates()
df1
```

```python
for a in range(len(df1)):
    try:
        df1['likes'][a] = df1['likes'][a].replace(',','')
    except:
        print(df1['likes'][a])
```

```python
for a in range(len(df1)):
    df1['likes'][a] = int(df1['likes'][a])
```

```python
df1.sort_values(by=['likes'], axis=0, ascending=False, inplace=True)
df1.head(7)
```

```python
df1.to_csv('./crawling_data/instagram_influencer.csv')
```

```python
df2 = pd.read_csv('./crawling_data/crawling_youtube_final.csv')
df2.drop(df2.columns[0], axis=1, inplace=True)
df2
```

```python
df2.rename(columns = {'youtuber': 'id'}, inplace = True)
df2.drop_duplicates(inplace=True)
df2['likes'] = df2['likes'].astype('int')
df2
```

```python
df2.drop_duplicates(subset='title',inplace=True)
```

```python
df2.reset_index(inplace=True, drop=True)
df2
```

```python
for i in range(len(df2)):
    df2['title'][i] =  df2['title'][i] + ' ' + df2['comment'][i]
df2
```

```python

df2.drop(df2.columns[3], axis=1, inplace=True)
df2
```
